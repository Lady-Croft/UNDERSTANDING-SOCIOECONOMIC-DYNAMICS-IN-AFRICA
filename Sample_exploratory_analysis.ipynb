```python
import pandas as pd

# Assuming df is your original DataFrame with a 'Country' column
# Replace this with your actual data loading or DataFrame creation process

# List of African countries
african_countries = ["Algeria", "Angola", "Benin", "Botswana", "Burkina Faso", "Burundi", "Cabo Verde", "Cameroon", "Central African Republic", "Chad", "Comoros", "Congo (Congo-Brazzaville)", "Côte d'Ivoire (Ivory Coast)", "Djibouti", "Egypt", "Equatorial Guinea", "Eritrea", "Eswatini", "Ethiopia", "Gabon", "Gambia", "Ghana", "Guinea", "Guinea-Bissau", "Kenya", "Lesotho", "Liberia", "Libya", "Madagascar", "Malawi", "Mali", "Mauritania", "Mauritius", "Morocco", "Mozambique", "Namibia", "Niger", "Nigeria", "Rwanda", "Sao Tome and Principe", "Senegal", "Seychelles", "Sierra Leone", "Somalia", "South Africa", "South Sudan", "Sudan", "Tanzania", "Togo", "Tunisia", "Uganda", "Zambia", "Zimbabwe"]

# Filter the DataFrame to include only African countries
df_africa_co2 = co2_data[co2_data['Country'].isin(african_countries)]

# Now, df_africa contains only records for African countries
```

## Exploratory Analysis on CO2 Production

```python
co2_data.head(10)
```

```python
display(df_africa_co2.head(10))
print(df_africa_co2.info())

print('Shape:', df_africa_co2.shape)

## Using Print before a query allows you get outputs for multiple queries in a single input cell
## Display also presents result more orderly than print
```

### View Percentage of Missing Data

```python
missing_percentage = (df_africa_co2.isna().sum() / len(co2_data)) * 100
rounded_percentage = missing_percentage.round(1) 
## the digit in the bracket is the number of decimal places, this approximates the value
sorted_percentage = rounded_percentage.sort_values(ascending=False)
print(sorted_percentage)
```

### Truncate Missing Data

```python
co2_data = df_africa_co2.dropna()
print(df_africa_co2.isna().sum())
print('Shape:', df_africa_co2.shape)
```

### View Columns in the Dataset

```python
co2_data.columns
```

### Rename Columns to Display just the Year

```python
co2_data.columns = [col.replace('co2_prod_', '') for col in co2_data.columns]
```

### Sum up CO2 Production for Each Year

```python
co2_sum = co2_data[['1990', '1991',
       '1992', '1993', '1994', '1995', '1996', '1997', '1998', '1999', '2000',
       '2001', '2002', '2003', '2004', '2005', '2006', '2007', '2008', '2009',
       '2010', '2011', '2012', '2013', '2014', '2015', '2016', '2017', '2018',
       '2019', '2020', '2021']].sum()
```

### Using the Sum Above Plot a Graph for Yearly Production of CO2 Across Countries

```python
plt.figure(figsize= (12,8))
sns.set_theme(font_scale=1.5)
sns.lineplot(co2_sum, x = co2_sum.index , y = co2_sum, color='firebrick')
plt.xticks(rotation = 45) 
## In a weird way changing the rotation value affects the appearance of the x-axis labels.

plt.grid(True, color = 'grey', linestyle = '--', alpha = 0.5)

plt.title('Annual Global CO2 Emission ', fontweight = 'bold')
plt.xlabel('Years', fontweight = 'bold')
plt.ylabel('Tonnes of carbon dioxide (CO2)', fontweight = 'bold')
```

### Plot a Graph to View the Production of CO2 in Each Country in 2021

```python
co2_2021_sorted = co2_data.sort_values(by = '2021', ascending = False).reset_index(drop=True)

plt.figure(figsize=(12, 8))
sns.barplot(x=co2_2021_sorted['Country'], y=co2_2021_sorted['2021'], color='turquoise')
plt.title('Co2 Production in 2021 by country')
plt.xlabel('Country', fontweight='bold')
plt.ylabel('Co2 Production in 2021', fontweight='bold')

# Add a vertical red dashed line after country 13 
plt.axvline(x=13.5, color='red', linestyle='--')

plt.grid(True, alpha=0.3, linestyle='--', axis='y')
plt.gca().set_xticklabels([])
plt.tight_layout()
plt.show()
```

```python
# Create a histogram for CO2 emissions in the year 2021
plt.figure(figsize=(8, 4))
sns.histplot(data=co2_data, x="2021")  # Use the correct column name for 2021
plt.xlabel("CO2 Emissions in 2021")
plt.ylabel("Frequency")
plt.title("Distribution of CO2 Emissions in 2021")
plt.show()
```
